### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def create_line_plot():
    try:
        # Generate x values and verify output
        x = np.linspace(0, 10, 100)
        if not isinstance(x, np.ndarray) or len(x) == 0:
            raise ValueError("Failed to generate valid x values with np.linspace")

        # Calculate y values and verify
        y = x**2
        if not isinstance(y, np.ndarray) or len(y) != len(x):
            raise ValueError("Failed to compute valid y values")

        # Check for invalid values (NaN or Inf)
        if not np.all(np.isfinite(x)) or not np.all(np.isfinite(y)):
            raise ValueError("Data contains invalid values (NaN or Inf)")

        # Create the plot
        plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)

        # Customize the plot
        plt.title('Line Plot of y = x^2', fontsize=14, pad=10)
        plt.xlabel('x values', fontsize=12)
        plt.ylabel('y values (x^2)', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend(loc='upper left')

        # Set axis limits
        plt.xlim(0, 10)
        plt.ylim(0, 100)

        # Adjust layout and show
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"Error: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Execute the function
if __name__ == "__main__":
    create_line_plot()
```
